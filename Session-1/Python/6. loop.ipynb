# Assignment 1: Write a script that processes a list of temperature readings. If any temperature is above a certain threshold, print a warning.

```python
# List of temperature readings
temperatures = [23, 35, 29, 41, 38, 30, 25]
threshold = 37

for temp in temperatures:
    if temp > threshold:
        print(f"⚠️ Warning: Temperature {temp}°C exceeds the threshold of {threshold}°C!")
    else:
        print(f"Temperature {temp}°C is within safe limits.")
```

# Assignment 2: Create a script that processes a dictionary of products, checking stock levels and generating restock alerts if necessary.

```python
# Product inventory
products = {
    "laptop": {"stock": 4, "min_required": 10},
    "smartphone": {"stock": 15, "min_required": 5},
    "tablet": {"stock": 2, "min_required": 7},
    "headphones": {"stock": 20, "min_required": 10}
}

# Process
for product, details in products.items():
    stock = details["stock"]
    min_required = details["min_required"]

    if stock < min_required:
        restock_amount = min_required - stock
        print(
            f"⚠️ Alert: {product.capitalize()} stock is low.\n"
            f"    Current stock: {stock}, Minimum required: {min_required}.\n"
            f"    Suggested restock: {restock_amount} units."
        )
    else:
        print(f"✅ {product.capitalize()} stock is sufficient. Current stock: {stock}.")
```
