# Assignment 1: Create a 2D list representing a 3x3 matrix and perform operations like accessing, modifying, and iterating through it.

```python
# The 3x3 Matrix
matrix_2D = [[47, 54, 62], [66, 70, 73], [46, 49, 64]]

# Accessing elements
print("Element of [2][1]:", matrix_2D[2][1])
# Modifying an element
matrix_2D[2][1] = 33
print("Modified Matrix:", matrix_2D)
# Iterating through the matrix
for row in matrix_2D:
    for element in row:
        print(element, end=" ")
    print()
```

# Assignment 3: Create a list of tuples, where each tuple contains a student's name and their grade. Sort this list by grades.

```python
students = [("Rima", 62), ("Sumaiya", 54), ("Jinia", 47), ("Saba", 73)]
sorted_students = sorted(students, key=lambda x: x[1])

print("Sorted Students by Grades:", sorted_students)
```
