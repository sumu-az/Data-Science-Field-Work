# Assignment 1: Create a dictionary representing a student with keys like 'name', 'roll_number', 'grades' (a list of subjects and marks).
# Perform various operations like adding, removing, and modifying elements.

```python
# Student Information
student = {
    'name': "Sumaiya",
    'roll_number': 201254,
    'grades': {'Math': 91, 'Science': 87, 'English': 90}
}

# Add a new subject
student['grades']['History'] = 80
# Modify an existing grade
student['grades']['Math'] = 95
# Remove a subject
del student['grades']['Science']

print("Updated Student Dictionary:", student)
```

# Assignment 2: Create a dictionary where keys are student names and values are lists of grades. Calculate the average grade for each student.

```python
grades_dict = {
    'Sumaiya': [95, 91, 89],
    'Jinia': [82, 90, 84],
    'Rima': [92, 87, 91]
}

average_grades = {name: sum(grades) / len(grades) for name, grades in grades_dict.items()}
print("Average Grades:", average_grades)
```
