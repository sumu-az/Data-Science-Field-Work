# Assignment 1: Write a function that calculates compound interest using the formula A = P(1 + r/n)^(nt).

```python
import math
def compound_interest(p, r, n, t):
    A = p * math.pow((1 + r / n), n * t)
    return A

principal = 5000
rate = 0.05
n = 5
t = 10

final_amount = compound_interest(principal, rate, n, t)
print(f"Compound Interest: BDT {final_amount:.2f} TK")
```

# Assignment 2: Create a script that prints the current time and updates every second.

```python
import time
import datetime

for _ in range(10):
    current_time = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    print("Current time:", current_time)
    time.sleep(1)

print("Done!")
```
