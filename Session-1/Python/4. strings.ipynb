# Assignment 1: Create a string that contains a simple bio data like name, age, and country. Extract each piece of information and print them separately.

```python
bio_data = "Name: Sumaiya, Age: 24, Country: Bangladesh"
name = bio_data.split(", ")[0].split(": ")[1]
age = bio_data.split(", ")[1].split(": ")[1]
country = bio_data.split(", ")[2].split(": ")[1]
print("Extracted Info:", f"Name: {name}, Age: {age}, Country: {country}")
```

# Assignment 2: Create a formatted string that includes data from a list or dictionary. For example, use a dictionary to store a person's information and format a string to include it.

```python
person_info = {"name": "Sumaiya", "age": 24, "country": "Bangladesh"}
formatted_info = f"{person_info['name']} is {person_info['age']} years old and lives in {person_info['country']}."
print("Formatted Info:", formatted_info)
```

# Assignment 3: Write a function that takes a string and returns a dictionary with the counts of each character in the string.

```python
def char_count(string):
    return {char: string.count(char) for char in set(string)}

str = "I'm learning strings!"
print("Character Count:", char_count(str))
```

# Assignment 4: Write a regex to find all the hashtags in a string.

```python
import re
text_with_hashtags = "This is a #great day to learn #regex in #Python!"
hashtags = re.findall(r"#\w+", text_with_hashtags)
print("Hashtags Found:", hashtags)
```

# Assignments: Write a regex to find the Bangladesh phone number with all variations.
# 01454565767
# +880196345634
# 0088785674657
# 01845-567567

```python
regex1 = r"\b(?:\+880|0088)?1[3-9]\d{8}\b|\b1[3-9]\d{2}-\d{6}\b"
regex2 = "Valid numbers: 01454565767, +880196345634, 0088785674657, 01845-567567."
regex3 = re.findall(regex1, regex2)
print("Bangladesh Phone Numbers Found:", regex3)
```
