# Assignment 1: Write a script that reads a CSV file containing product information and converts it into a JSON file.

```python
import csv
import json

from google.colab import files
uploaded = files.upload()
csv_file_path = next(iter(uploaded))

json_file_path = csv_file_path.replace('.csv', '.json')

with open(csv_file_path, 'r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    products = [row for row in csv_reader]

with open(json_file_path, 'w') as json_file:
    json.dump(products, json_file, indent=4)

print(f"CSV data has been converted to JSON and saved to {json_file_path}.")
```

# Assignment 2: Create a log file writer that appends log messages to a file with timestamps.

```python
from datetime import datetime
from google.colab import files

uploaded = files.upload()
log_file_path = next(iter(uploaded))

def log_message(message):
    timestamp = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
    log_entry = f"{timestamp} - {message}\n"

    # append mode
    with open(log_file_path, 'a') as file:
        file.write(log_entry)

log_message("Application started.")
log_message("Processing CSV to JSON conversion.")
log_message("Application finished.")

with open(log_file_path, 'r') as file:
    print("Log file content:")
    print(file.read())
```
